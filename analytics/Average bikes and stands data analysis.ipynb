## Find averages

```python
# Create new dataframe for hourly available bikes
df_all = pd.DataFrame(columns=['hour','weekday','a_bikes'])
df_all.head(5)
```

```python
# Group by station number and find the mean
x = 1
while x <= 105:
    res3 = df.loc[df['number'] == x]
    res3a = res3.groupby(['hour', 'weekday'])['a_bikes'].mean().reset_index()
    res3a['number'] = x
    df_all = df_all.append(res3a)
    x+=1
df_all.shape
```

```python
df_all.head(5)
```

```python
# Create new dataframe for daily available bikes
df_all_day = pd.DataFrame(columns=['weekday','a_bikes'])
df_all_day.head(5)
```

```python
# Group by station number and find the mean
j = 1
while j <= 105:
    res5 = df.loc[df['number'] == j]
    res5a = res5.groupby(['weekday'])['a_bikes'].mean().reset_index()
    res5a['number'] = j
    df_all_day = df_all_day.append(res5a)
    j+=1
df_all_day.shape
```

```python
df_all_day.head(5)
```

```python
# Create new dataframe for hourly available stands
df_all2 = pd.DataFrame(columns=['hour','weekday','a_stands'])
```

```python
# Group by station number and find the mean
y = 1
while y <= 105:
    res4 = df.loc[df['number'] == y]
    res4a = res4.groupby(['hour', 'weekday'])['a_stands'].mean().reset_index()
    df_all2 = df_all2.append(res4a)
    y+=1
df_all2.shape
```

```python
df_all2.head(5)
```

```python
# Create new dataframe for daily available stands
df_all_day2 = pd.DataFrame(columns=['weekday','a_stands'])
df_all_day2.head(5)
```

```python
# Group by station number and find the mean
k = 1
while k <= 105:
    res6 = df.loc[df['number'] == k]
    res6a = res6.groupby(['weekday'])['a_stands'].mean().reset_index()
    df_all_day2 = df_all_day2.append(res6a)
    k+=1
df_all_day2.shape
```

```python
df_all_day2.head(5)
```

```python
# Reset the index for concat
df_all.reset_index()
```

```python
df_all_day.reset_index()
```

```python
# Change station numbers to an int
df_all.number = df_all.number.astype(int)
```

```python
df_all_day.number = df_all_day.number.astype(int)
```

```python
# Reorder columns
df_all = df_all[['number', 'weekday', 'hour', 'a_bikes']]
df_all.head(5)
```

```python
df_all_day = df_all_day[['number', 'weekday', 'a_bikes']]
df_all_day.head(5)
```

```python
df_all2.reset_index()
```

```python
df_all_day2.reset_index()
```

```python
# Create final df with stands column added
df_final = pd.concat([df_all, df_all2[['a_stands']]], axis=1)
df_final.head(5)
```

```python
df_final2 = pd.concat([df_all_day, df_all_day2[['a_stands']]], axis=1)
df_final2.head(5)
```

```python
# Round each average float and then turn into an int
df_final.a_bikes = df_final.a_bikes.round(0)
df_final.a_stands = df_final.a_stands.round(0)
df_final.a_bikes = df_final.a_bikes.astype(int)
df_final.a_stands = df_final.a_stands.astype(int)
df_final.shape
```

```python
df_final2.a_bikes = df_final2.a_bikes.round(0)
df_final2.a_stands = df_final2.a_stands.round(0)
df_final2.a_bikes = df_final2.a_bikes.astype(int)
df_final2.a_stands = df_final2.a_stands.astype(int)
df_final2.shape
```
